```python
df = pd.read_csv('myntra_products_catalog.csv')
df[:2]
```

```python
df = df.replace({'Boys':'Men', 'Girls':'Women'})
df = df[df['Gender']!='Unisex Kids']

df = df.replace({'U.S. Polo Assn. Kids':'US Polo'})
```

```python
# Replacing null values to Missing
df['PrimaryColor'] = df['PrimaryColor'].replace(np.nan,"Missing")
```

```python
df[:2]
```

```python
fig = px.pie(df, names='Gender', 
             height=300, width=600, 
             title='Gender Overview',
             color_discrete_sequence=['#4c78a8', '#72b7b2', '#6b92bc'])
fig.show()
```

```python
fig = px.pie(df, names='Gender', 
             height=300, width=600, 
             hole=0.7, 
             title='Gender Overview',
             color_discrete_sequence=['#4c78a8', '#72b7b2', '#6b92bc'])
fig.update_layout(plot_bgcolor='#fafafa', paper_bgcolor='#fafafa')
fig.show()
```
